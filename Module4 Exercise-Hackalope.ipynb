1. How many occurences of "row" is in "Row, row, row your boat. Gently down the stream. Merrily, merrily, merrily, merrily. Life is but a dream. Row, row, row your boat. Gently up the creek."

```python
import re

dat_string = 'Row, row, row your boat. Gently down the stream. Merrily, merrily, merrily, merrily. Life is but a dream. Row, row, row your boat. Gently up the creek.'

rows = re.findall("[Rr]ow", dat_string)

print(f'There are {dat_string.count("row")} using string methods, and {len(rows)} matches using regex in the string:\n\n\"{dat_string}\"')
```

6. Convert 39 to a binary number using floor division and mod operators. Then, compare your answer using bin() and int() functions.

```python
base = 2
num_in = 39

out = str()

while num_in > 0:
  out = str(num_in % base) + out
  num_in = num_in // base

print(out)
```

```python
bin(int(39))
```
