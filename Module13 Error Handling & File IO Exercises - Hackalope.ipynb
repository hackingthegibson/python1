### Exercise 5
Write a program that asks the user for a number and a file name, then prints out the number from 1 to that number, one one each line to a file named using user's input.
Then, read that file and print out its content.

```python
def write_num_file(filename, num):
  try:
    fh = open(filename, "w")
    for line in range(1,num+1):
      fh.write(f"{str(line)}\n")
    fh.close()
    retval=True
  except:
    retval=False
  return retval

def with_write_num_file(filename, num):
  try:
    with open(filename, "w") as fh:
      for line in range(1,num+1):
        fh.write(f'{str(line)}\n')
    retval=True
  except:
    retval=False
```

```python
get_name = input("Provide a file name: ")
get_num = int(input("How many lines do you want? "))

foo = with_write_num_file(get_name, get_num)

with open(get_name) as fh:
  print(fh.read())
```

### Exercise 6
Merge a file from exercise 5 and <b>exercise1.txt</b> to a file called <b>merged.txt</b>. Then, prin out the content of <b>merged.txt</b> before closing the file.

```python
files = [ get_name, "exercise1.txt"]

def file_merge(filename, file_list):
  with open(filename, "w") as merge_fh:
    for file in file_list:
      with open(file) as fh:
        merge_fh.write(fh.read())

file_merge("merged.txt", files)
```

```python
with open("merged.txt") as fh:
  print(fh.read())
```
