### Exercise 6.
Create a Test Scores program that gets the name of students and test score from a user. Then, generate a list of students who passed and a list of students who failed where passing test score is 60 or above.

```python
students = {}

while True:
  name = input("Student Name: ")
  if name == "" :
    break
  score = float(input("Score: "))
  students[name] = score
```

```python
print("\nThese students passed (70 or higher)")

for name, score in students.items():
  if score >= 60:
    print(name.title())

print("\nThese students passed (Lower than 70)")
for name, score in students.items():
  if score < 70:
    print(name.title())
```

### Exercise 7.

Create a Word Counter program that gets a string of text from a user and then counts the number of times a word occrs in the string.

The output should look something like this:

    The Word Counter Program
    
    Enter a text string:
    Humpty Dumpty sat on a wall, Humpty Dumpty had a great fall. All the king’s horses and all the king’s men couldn’t put Humpty together again.

    a : 2
    again : 1
    all : 2
    and : 1
    couldn’t : 1
    dumpty : 2
    fall : 1
    great : 1
    had : 1
    horses : 1
    humpty : 3
    king’s : 2
    men : 1
    on : 1
    put : 1
    sat : 1
    the : 2
    together : 1
    wall : 1

    There were total 26 words in the text string   

```python
# big_string = input("Enter a text string: ")
big_string = 'Humpty Dumpty sat on a wall, Humpty Dumpty had a great fall. All the king’s horses and all the king’s men couldn’t put Humpty together again.'

big_string = big_string.lower()

words = big_string.split()

print(words)
```

```python
counter = {}

for word in words:
  word = word.strip(".,")
  if word in counter:
    counter[word] += 1
  else:
    counter[word] = 1

for word, count in sorted(counter.items()):
  print(f'{word}: {str(count)}')
```
