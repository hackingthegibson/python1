```python
from random import randint

rps = [ "rock", "paper", "scissors"]
spr = [ "scissors", "paper", "rock"]

def cheat(throw):
  cheat = spr.index(throw) - 1
  return("I choose " + spr[cheat] + ' - I Win!')

def fair(throw):
  prog_try = randint(0,len(rps)-1)
  throw_index = rps.index(throw)

  if throw_index - prog_try == 0:
    out="I choose " + rps[prog_try] + ' - Tie!'
  elif prog_try - throw_index == 1:
    out="I choose " + rps[prog_try] + ' - I Win!'
  else:
    out="I choose " + rps[prog_try] + ' - I Lose'
  return out

player_try = input("rock, paper, sissors shoot! ")
print()
print(cheat(player_try))
print(fair(player_try))
```
