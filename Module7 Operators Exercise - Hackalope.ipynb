### Exercise 3

Ask a user to enter a phrase. If the word contains a number, display the "Number found" message. Otherwise, show the "Number not found" message.

```python
phrase = input("Gimme some input : ")

check = False

for char in phrase:
  if char.isdigit():
    check = True

if check:
  print("Number found")
else:
  print("Number not found")
```

```python
import re

phrase = input("Gimme some input : ")

if re.search(r'\d+',phrase):
  print("Number found")
else:
  print("Number not found")
```

### Exercise 4

Ask a user to enter a number. Then, only if the number is divisible by 2 and 3, show the result of the number divided by two and three in a formated string. The output should look something like this:
        
    Enter a number: 2
    Bye    
    
    Enter a number: 60
    60 / 2 = 30.0
    60 / 3 = 20.0
    Bye
    
    Enter a number: 3
    Bye
    
    Enter a number: 10
    Bye    

```python
val = int(input("Enter a number : "))

if not val % 2 and not val % 3:
  print(f'{val} / 2 {(val / 2):.1f}')
  print(f'{val} / 3 {(val /3):.1f}')

print("Bye")
```

### Exercise 5
