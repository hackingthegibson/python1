### Exercise 1
Ask a user for 5 names and place each one in to a list called <i>names</i>. Then, print <i>names</i>.

```python
def num_suffix(x):
  f_digit = int(str(x)[-1])
  if f_digit == 1:
    return(f'{x}st')
  elif f_digit == 2:
    return(f'{x}nd')
  elif f_digit == 3:
    return(f'{x}rd')
  else:
    return(f'{x}th')

num_names = 5
names = []
for i in range(0,num_names):
  names.append(input(f'Give me the {num_suffix(i+1)} name : '))

print()
print(names)
```

### Exercise 2
Print out a randomly selected name in the names list from exercise 1.

```python
import random as rand

print('A random name is {}'.format(names[rand.randint(0,num_names-1)]))
```

### Exercise 4
Finish writing a program to generate 10 random numbers between 1 and 100 and save it to a list.
Finally, print out the original numbers followed by the smallest, largest, and average number.

```python
import random as rand

numbers = []
total = 0

for i in range(1, 11) :
    numb = rand.randint(1, 100)
    total += numb
    numbers.append(numb)


# some code needed here to find the smallest, largest, and average number

avg = total / len(numbers)

# print("Total {} : Items {} : Mean {}".format(total, len(numbers), avg))

print(f"Org Numbers are {numbers}")
numbers.sort()
print(f"Sorted numbers are {numbers}")
print(f"Largest number is {numbers[-1]}")
print(f"Smallest number is {numbers[0]}")
print(f"Average is {avg}")
```

### Exercise 5

Write a function called <b>draw_triangle()</b> to draw a right triangle where the base of the triangle is the number passed in as an arguement. Call <b>draw_triangle()</b> to show it works.

e.g. <b>draw_triangle(7)</b> draws as below:

    *
    **
    ***
    ****
    *****
    ******
    *******


```python
# write code for draw_triangle() function

def draw_triangle(height):
  for i in range(1,height+1):
    print("*" * i)

def draw_center_triangle(height):
  for i in range(0,height):
    pad = height - i
    layer = (i*2) + 1
    print(" " * pad, "*" * layer)

def draw_invert_center_triangle(height):
  for i in range(height-1,-1,-1):
    pad = height - i
    layer = (i*2) + 1
    print(" " * pad, "*" * layer)
```

```python
base = int(input("Enter a number: "))

draw_triangle(base)
print()
draw_center_triangle(base)
print()
draw_invert_center_triangle(base)
```
