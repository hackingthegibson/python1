### Exercise 2
Ask a user to enter two numbers and find their average. Then, print out the average and whether each number is below or above average. Determine if a function should be created or not.
    

```python
def above_below(num, avg):
  if num > avg:
    out = "above"
  else:
    out = "below"
  return out
```

```python
foo = int(input("Enter number : "))
bar = int(input("Enter number : "))

avg = (foo + bar) / 2
```

```python
print('The average is {:,.2f}.\nThe number {} is {} the average.\nThe number {} is {} the average.'.format(avg, foo, above_below(foo, avg), bar, above_below(bar, avg)))
```

### Exercise 3
Write a function to convert a test score to a grade where  
 - 90 - 100: A
 - 80 - 89:  B
 - 70 - 79:  C
 - 60 - 69:  D
 - < 60   :  F

If a score is not between 0 and 100 then print out an error message. In the program, ask the user for a score and then call your function.

```python
score = int(input("Submit score : "))

if score > 100:
  print("Error")
elif score >= 90:
  print("A")
elif score >= 80:
  print("B")
elif score >= 70:
  print("C")
elif score >= 60:
  print("D")
elif score < 60 and score >= 0:
  print("F")
else:
  print("Error")

```

### Exercise 5
When a = 61 and b = 15, what is the result of following bitwise opertion? Use inline comments to show your result in a binary number. Find your answer by hand.

    a & b
    a | b
    a ^ b
    a >> 2
    

```python
a = 61
b = 15

print('the and (&) operation is {} and is bin {:b}'.format(a & b, a & b))
print('the or (|) operation is {} and is bin {:b}'.format(a | b, a | b))
print('the xor (^) operation is {} and is bin {:b}'.format(a ^ b, a ^ b))
print('the shift a 2 (>> 2) operation is {} and is bin {:b}'.format(a >> 2, a >> 2))
```
