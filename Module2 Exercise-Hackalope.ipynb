#### 2. How much water runs off a roof in a rain storm?

Multiply the roof dimensions by the number of inches of rainfall. As an example, 600" x 240" x 1" = 144,000 cubic inches of water for an inch of rainfall. Divide that result by 231 to get the number of gallons (because 1 gallon = 231 cubic inches). (144,000/231 = 623.38).

> LOOP **in_rain** from 1 to 10 by adding 1

>> SET **runoff** TO **in_rain** x **runoff_factor**

>> PRINT "**in_rain** produces **runoff**"

> END LOOP

END
